### List of employees wth job titles & department names:

```python
%sql SELECT e.emp_name, j.job_title, d.dept_nm \
     FROM Employee e \
     JOIN Employment_History ehist \
     ON e.emp_id = ehist.emp_id \
     JOIN Job j \
     ON ehist.job_title_id = j.job_title_id \
     JOIN Department d \
     ON ehist.dept_id = d.dept_id \
     WHERE end_dt IS NULL;
```

### Number of Employees in each department:

```python
%sql SELECT d.dept_nm, COUNT(*) AS employee_count \
     FROM Employee e \
     JOIN Employment_History ehist \
     ON e.emp_id = ehist.emp_id \
     JOIN Department d \
     ON d.dept_id = ehist.dept_id \
     WHERE end_dt IS NULL \
     GROUP BY d.dept_nm;
```

### Current & past jobs for employee Toni Lembeck:
(Including emp_name, job_title, dept, mngr_name, start & end_dt)

```python
%sql SELECT e.emp_name, j.job_title, d.dept_nm, m.manager, start_dt, end_dt \
     FROM Employee e \
     JOIN Employment_History ehist \
     ON e.emp_id = ehist.emp_id \
     JOIN Job j \
     ON ehist.job_title_id = j.job_title_id \
     JOIN Department d \
     ON ehist.dept_id = d.dept_id \
     JOIN Manager m \
     ON m.manager_id = ehist.manager_id \
     WHERE e.emp_name = 'Toni Lembeck';
```
